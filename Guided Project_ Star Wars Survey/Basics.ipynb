First of all, I'll only consider those who watched all of the movies (the previous star_wars_watchall subset I've created, renaming it to swwa for easier typing).

```python
swwa = star_wars_watchall
for c in swwa.iloc[:,1:].columns:
    print(c)
    print(swwa[c].unique())
```

```python
#Checking if ratings were given correctly
ratings = {}
for i in range(1,7):
    s = 'rank_ep_' + str(i)
    for j in range(1,7):
        aux = len(swwa[swwa[s]==j].loc[:,s])
        rate = str(j)
        if rate not in ratings:
            ratings[rate] = aux
        else:
            ratings[rate] += aux
ratings
```

```python
#They were, except for the missing value on one episode 3. Can be treated. Treating it.
swwa['rank_ep_3'] = swwa['rank_ep_3'].fillna(6)
```

```python
swwa.iloc[:,9:15].info()
```

```python
ep_names = ['Episode 1','Episode 2','Episode 3','Episode 4','Episode 5','Episode 6']
swwa_trekker = swwa.groupby('is_trekker')
swwa_trekker = swwa_trekker.mean()
swwa_trekker
```

```python
swwa_trekker_yes_rankings = swwa_trekker.iloc[0,8:14]
swwa_trekker_yes_rankings
```

```python
swwa_trekker_no_rankings = swwa_trekker.iloc[1,8:14]
swwa_trekker_no_rankings
```

```python
fig, ax = plt.subplots()
ind = arange(len(ep_names))+1
w = 0.5
rects1 = ax.bar(ind - w/4, swwa_trekker_yes_rankings, w/2)
rects2 = ax.bar(ind + w/4, swwa_trekker_no_rankings, w/2)
tick_positions = range(1,len(ep_names)+1)
ax.set_xticks(tick_positions)
ax.set_xticklabels(ep_names)
ax.set_title("Rankings of each of Star Wars Episode \n (only prequels and originals) \n Grouped by Star Trek fans or not")
ax.legend((rects1[0], rects2[0]), ('Fan', 'Not Fan'))
plt.plot()
```

Trekker fans seem to rate prequels slightly better, but nothing very conclusive.

```python
swwa_education = swwa.groupby('education')
swwa_education = swwa_education.mean()

education_names = ['High school degree', 'Some college or Associate degree', 'Bachelor degree', 'Graduate degree', 'Less than high school degree']
count_types = len(education_names)

fig, ax = plt.subplots()
ind = arange(len(ep_names))+1
w = 0.5
swwa_education_rankings = swwa_education.iloc[:,8:14]
print(swwa_education_rankings)

for i, row in enumerate(swwa_education_rankings.values):
    ax.bar(ind - w + 2*(i+1)*w/(count_types+1), row, 2*w/(count_types+1))
tick_positions = range(1,len(ep_names)+1)
ax.set_xticks(tick_positions)
ax.set_xticklabels(ep_names)
ax.set_title("Rankings of each of Star Wars Episode \n (only prequels and originals) \n Grouped by Education Degree")
ax.legend(education_names, loc=8, bbox_to_anchor=(0.25, -0.5, 0.5, 1))
plt.plot()
```

High school and college/associare follows the common pattern.
Bachelor degrees likes prequels a little better than previous ones.
Graduate degree seems to be composed of very little sample size.
Less than high school degree has some more liking for Episode 1 specifically.

```python
swwa_location = swwa.groupby('location')
swwa_location = swwa_location.mean()

location_names = ['South Atlantic', 'West North Central', 'Middle Atlantic', 'East North Central', 'Mountain', 'West South Central', 'Pacific', 'New England', 'East South Central']
count_types = len(location_names)
swwa_location_rankings = swwa_location.iloc[:,8:14]
print(swwa_location_rankings)

fig, ax = plt.subplots()
ind = arange(len(ep_names))+1
w = 0.5

for i, row in enumerate(swwa_location_rankings.values):
    ax.bar(ind - w + 2*(i+1)*w/(count_types+1), row, 2*w/(count_types+1))
tick_positions = range(1,len(ep_names)+1)
ax.set_xticks(tick_positions)
ax.set_xticklabels(ep_names)
ax.set_title("Rankings of each of Star Wars Episode \n (only prequels and originals) \n Grouped by Location")
ax.legend(location_names, loc=8, bbox_to_anchor=(0.25, -0.8, 0.5, 1))
plt.plot()
```
